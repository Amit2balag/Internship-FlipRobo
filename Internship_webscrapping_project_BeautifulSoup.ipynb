***6. Write a python program to scrape cricket rankings from ‘www.icc-cricket.com’.

*Top ODI teams

```python
page = requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page
soup = BeautifulSoup(page.content) # fetched the content in the webpage
soup.prettify
soup
team = soup.find_all(class_='u-hide-phablet')
team
Team = []
for i in team:
  if i.text is None:
    Team.append('No_data')
  else:
    Team.append(i.text.replace('\n', ""))
Team
#len(Team)
```

```python
Team = Team[0:10]
#len(Team)
Team
```

```python
Top_team_matches = soup.find_all('td', class_='rankings-block__banner--matches')
Top_team_matches
Top_Team_Matches = []
for i in Top_team_matches:
    Top_Team_Matches.append(i.text.strip())
#len(Top_Team_Matches)
Top_Team_Matches
```

```python
matches = soup.find_all('td', class_='table-body__cell u-center-text')
matches
Matches = []
for i in matches:
    Matches.append(i.text.strip())
#len(Matches)
Matches
```

```python
del Matches[1::2]
Matches
```

```python
Matches = Top_Team_Matches+Matches
Matches
```

```python
points = soup.find_all('td', class_='table-body__cell u-center-text')
points
Points = []
for i in points:
    Points.append(i.text.strip())
Points
#len(Points)
Points
```

```python
Points = [v for i, v in enumerate(Points) if i % 2 != 0]
Points
```

```python
Top_team_points = soup.find_all('td', class_='rankings-block__banner--points')
Top_team_points
Top_team_Points = []
for i in Top_team_points:
    Top_team_Points.append(i.text.strip())
Top_team_Points
#len(Top_team_Points)
Top_team_Points
```

```python
Team_points = Points+Top_team_Points
Team_points
```

```python
rating = soup.find_all('td', class_='table-body__cell u-text-right rating')
rating
Rating = []
for i in rating:
    Rating.append(i.text.strip())
#len(Rating)
Rating
```

```python
Top_team_rating = soup.find_all('td', class_='rankings-block__banner--rating u-text-right')
Top_team_rating
Top_team = []
for i in Top_team_rating:
    Top_team.append(i.text.strip())
#len(Rating)
Top_team
```

```python
Ratings = Top_team + Rating
Ratings
```

```python
Top10_ODI_teams = pd.DataFrame({})
Top10_ODI_teams['Team'] = Team
Top10_ODI_teams['Matches'] = Matches
Top10_ODI_teams['Team_points'] = Team_points
Top10_ODI_teams['Ratings'] = Ratings
Top10_ODI_teams
```
